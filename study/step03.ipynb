```python
import numpy as np
```

```python
def multivariate_normal(x, mu, cov):
    det = np.linalg.det(cov)
    inv = np.linalg.inv(cov)
    D = len(x)
    z = 1 / np.sqrt((2 * np.pi) ** D * det)
    y = z * np.exp((x - mu).T @ inv @ (x - mu)/ -2.0)
    return y
```

```python
x = np.array([[0], [0]])
mu = np.array([[1], [2]])
cov = np.array([[1, 0],
                [0, 1]])
y = multivariate_normal(x, mu, cov)
y
```

```python
np.random.seed(0)

N = 10000
D = 2
xs = np.random.rand(N, D)

mu = np.sum(xs, axis=0)
mu /= N

cov = 0

for n in range(N):
    x = xs[n]
    z = x - mu
    z = z[:, np.newaxis]
    cov += z @ z.T

cov /= N
mu, cov
```
