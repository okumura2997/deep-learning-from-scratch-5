```python
import numpy as np
```

```python
def normal(x, mu=0, sigma=1):
    y = 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-(x - mu) ** 2 / (2 * sigma ** 2))
    return y
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 100)
y = normal(x)
```

```python
plt.plot(x, y)
```

```python
x = np.linspace(-10, 10, 1000)
y0 = normal(x, mu=-3)
y1 = normal(x, mu=0)
y2 = normal(x, mu=5)

plt.plot(x, y0, label='$\mu$=-3')
plt.plot(x, y1, label='$\mu$=0')
plt.plot(x, y2, label='$\mu$=5')
plt.legend()
```

```python
x = np.linspace(-10, 10, 1000)
y0 = normal(x, sigma=0.5)
y1 = normal(x, sigma=1)
y2 = normal(x, sigma=2)

plt.plot(x, y0, label='$\sigma$=0.5')
plt.plot(x, y1, label='$\sigma$=1')
plt.plot(x, y2, label='$\sigma$=2')
plt.legend()
```

```python
x_means = []

N = 1
for _ in range(10000):
    xs = []
    for n in range(N):
        x = np.random.rand()
        xs.append(x)
    mean = np.mean(xs)
    x_means.append(mean)
    
plt.hist(x_means, bins='auto', density=True)
plt.xlim(-0.05, 1.05)
plt.ylim(0, 5)
plt.show()
```

```python
x_sums = []
N = 5

for _ in range(10000):
    xs = []
    for n in range(N):
        x = np.random.rand()
        xs.append(x)
    t = np.sum(xs)
    x_sums.append(t)

x_norm = np.linspace(-5, 5, 1000)
mu = N / 2
sigma = np.sqrt(N / 12)
y_norm = normal(x_norm, mu, sigma)

plt.hist(x_sums, bins='auto', density=True)
plt.plot(x_norm, y_norm)
plt.xlim(-1, 6)
plt.show()
```
